```python
# Loại bỏ giá trị ngoại lai
from sklearn.preprocessing import RobustScaler as RS

#Lọc ngoại lai của price
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3-Q1
data_price = data[~((data['price'] < (Q1 - 1.5 * IQR)) | (data['price'] > (Q3 + 1.5 * IQR)))]

#Lọc ngoại lai của sqft_living
Q1 = data_price['sqft_living'].quantile(0.25)
Q3 = data_price['sqft_living'].quantile(0.75)
IQR = Q3-Q1
data_price = data_price[~((data_price['sqft_living'] < (Q1 - 1.5 * IQR)) | (data_price['sqft_living'] > (Q3 + 1.5 * IQR)))]

import seaborn as sns
import numpy as np

X = data_price.loc[:,'sqft_living']
Y = data_price.loc[:,'price']

X = np.asarray(X)
Y = np.asarray(Y)

w1 = (X.mean()*Y.mean()- (X*Y).mean())/(X.mean()*X.mean()-(X*X).mean())
w0 = Y.mean()-w1*X.mean()
Y_ = w0 + w1*X
sns.scatterplot(x = X, y = Y)
sns.lineplot(x=X,y= Y_)
```
